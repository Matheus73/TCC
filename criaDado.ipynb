```python
import pandas as pd
```

```python
df = pd.read_csv("dataset/ner_dataset_tratado.csv", encoding="latin1")
df.head()
```

```python
def sentence_integrate(data):
    agg_func = lambda s: [
        (w, p, t)
        for w, p, t in zip(
            s["Word"].values.tolist(),
            s["POS"].values.tolist(),
            s["Tag"].values.tolist(),
        )
    ]
    return data.groupby("Sentence").apply(agg_func).tolist()
```

```python
sentences = sentence_integrate(df)
sentences[0]
```

```python
len_sentences = len(sentences)
print("O valor total de sentenças é: ", len_sentences)
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.2, random_state=1, shuffle=False)
df_test.head(30)
```

```python
df_train.head(30)
```

```python
test_size = int(len_sentences * 0.1)
print("O valor de 10% das sentenças é: ", test_size)
resto = len_sentences - test_size
print("O valor de 90% das sentenças é: ", resto)
div_index = df.query(f"Sentence == 'Sentence: {test_size}'").tail(1).index[0]
print("O valor do index de divisão é: ", div_index)
```

```python
df_test = df.iloc[:div_index + 1] # 9591
df_resto = df.iloc[div_index + 1:] # 38367
df_test.tail()
```

```python
df_resto.tail()
```

```python
df_test.to_csv("dataset/ner_dataset_test.csv", index=False)
```

```python
div_index_train = df.query(f"Sentence == 'Sentence: {resto}'").tail(1).index[0]
print("O valor do index de divisão é: ", div_index_train)
```

```python
df_train = df.iloc[:div_index_train + 1]
df_unlabeled = df.iloc[div_index_train + 1:]
df_train.tail()
```

```python
df_unlabeled.head()
```

```python
df_unlabeled.to_csv("dataset/ner_dataset_unlabeled.csv", index=False)
df_train.to_csv("dataset/ner_dataset_train.csv", index=False)
```

```python
sentences_train = sentence_integrate(df_train)
sentences_test = sentence_integrate(df_test)
sentences_unlabeled = sentence_integrate(df_unlabeled)

print("O valor total de sentenças de treino é: ", len(sentences_train))
print("O valor total de sentenças de teste é: ", len(sentences_test))
print("O valor total de sentenças não rotuladas é: ", len(sentences_unlabeled))
```

```python
def get_num_words(data):
    words = list(set(data["Word"].values))
    words.append("ENDPAD")
    num_words = len(words)
    print("num_words ---------- ", num_words)

get_num_words(df_train)
get_num_words(df_test)
get_num_words(df_unlabeled)
get_num_words(df)
```
